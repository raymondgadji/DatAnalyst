##  **Visualisation de données**


 - Maîtriser les techniques de visualisation pour l'analyse exploratoire.
 - Savoir interpréter les visualisations pour identifier les patterns et les tendances.

**Principes Clés :**

    - Clarté :

Les graphiques doivent être clairs et faciles à comprendre. Évitez les éléments superflus et concentrez-vous sur l'information pertinente.

    - Pertinence :

Choisissez le type de graphique approprié en fonction du type de données et des questions que vous souhaitez explorer.

    - Simplicité :

Un graphique simple est souvent plus efficace qu'un graphique complexe. Limitez le nombre de couleurs et de types de lignes.

    - Précision :

Assurez-vous que les graphiques représentent fidèlement les données sans distorsion.

    - Consistance :

Utilisez des couleurs et des styles de manière cohérente pour faciliter la comparaison et la compréhension.

**Bonnes Pratiques :**

- Choisir le bon type de graphique en fonction des données et des objectifs de l'analyse.
- Étiqueter les axes et fournir une légende claire.
- Utiliser des couleurs avec discernement pour améliorer la lisibilité, mais éviter les palettes qui pourraient induire en erreur.
- Inclure des titres descriptifs qui expliquent ce que montre le graphique.
- Vérifier la précision des données et éviter les manipulations qui pourraient fausser les interprétations.


### Outils de Visualisation : Matplotlib, Seaborn..

**Matplotlib  :**
Matplotlib est une bibliothèque de visualisation en Python qui permet de créer des graphiques statiques, animés et interactifs.

Elle est très flexible et puissante, mais peut nécessiter du code plus détaillé pour les graphiques complexes.

https://matplotlib.org/

**Seaborn :**

Seaborn est construit sur Matplotlib et fournit une interface de haut niveau pour dessiner des graphiques attractifs et informatifs.

Il est conçu pour simplifier la création de graphiques complexes avec moins de code.

https://seaborn.pydata.org/examples/index.html

```python
#Importer les libraries de dépendances:
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Chargement des données
df = sns.load_dataset('titanic')
df.head()
```

### Types de Graphiques
#### Histogrammes :

```python
#Montrer la distribution des données continues.
plt.hist(df['age'].dropna(), bins=20, edgecolor='black')
plt.title('Distribution de l\'Âge')
plt.xlabel('Âge')
plt.ylabel('Fréquence')
plt.show()
```

#### boxplot

```python
# Montrer la distribution des données, y compris les quartiles et les valeurs aberrantes. 
sns.boxplot(x='class', y='age', data=df)
plt.title('Répartition de l\'Âge par Classe')
plt.show()
```

#### Scatter Plots :

```python
# Visualiser la relation entre deux variables continues.
plt.scatter(df['age'], df['fare'])
plt.title('Âge vs Fare')
plt.xlabel('Âge')
plt.ylabel('Fare')
plt.show()
```

#### Heatmap

```python
#Visualiser les données matricielles avec des variations de couleur.
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap des Corrélations')
plt.show()
```

#### Pair Plots :

```python
# Visualiser les relations entre toutes les paires de variables dans un dataset.
sns.pairplot(df[['age', 'fare', 'sibsp', 'parch']].dropna())
plt.title('Pair Plot des Variables Sélectionnées')
plt.show()
```
