#### Récupérer les livres de toutes les pages  

```python

# Fonction pour extraire les informations d'une page donnée
def extract_book_info(page_url):
    books=[]
    response = requests.get(page_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        #livres = soup.find_all(class_='product_pod')
        livres = soup.find_all('article', {'class': 'product_pod'})
        #print(livres)
        for livre in livres:

            titre = livre.h3.a.get('title')

            rating = livre.find('p', {'class': 'star-rating'}).get('class')[1]

            prix = livre.select('div p.price_color')[0].text[2:]

            disponibilite = livre.select('div p.availability')[0].text.strip()

            books.append({'title': titre, 'rating': rating, 'price': prix, 'availability': disponibilite})
        df_books=pd.DataFrame(books)
            
            
    return df_books

```

#### Créer une liste des livres avec chacune des catégories

```python
# Fonction pour extraire les titres des livres d'une page donnée
def extract_book_titles_from_page(url):
    response = requests.get(url)
    if response.status_code != 200:
        return []
    soup = BeautifulSoup(response.content, 'html.parser')
    books = soup.find_all('article', class_='product_pod')
    titles = [book.h3.a.get('title') for book in books]
    return titles

# Fonction pour extraire tous les livres d'une catégorie
def scrape_category(category_name, category_url):
    book_titles = []
    page_number = 1
    while True:
        page_url = f"{category_url.replace('index.html', '')}page-{page_number}.html"
        titles = extract_book_titles_from_page(page_url)
        if not titles:  # Si aucune donnée, la page n'existe pas, arrêter
            break
        book_titles.extend(titles)
        page_number += 1
    return category_name, book_titles, len(book_titles)

# URL de la page principale
url = "https://books.toscrape.com/index.html"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Trouver la barre latérale des catégories
sidebar = soup.find('div', class_='side_categories')
category_links = sidebar.find_all('a')

# Liste pour stocker les résultats
categories_data = []

# Parcourir les catégories (ignorer la catégorie générale "Books")
for link in category_links:
    if 'books_1' not in link['href']:
        category_name = link.text.strip().split('(')[0].strip()
        category_url = f"https://books.toscrape.com/{link['href']}"
        name, titles, count = scrape_category(category_name, category_url)
        categories_data.append({
            'Category': name,
            'Number of Books': count,
            'Book Titles': titles
        })

# Créer un DataFrame
df_categories = pd.DataFrame(categories_data)

# Afficher le DataFrame
print(df_categories[['Category', 'Number of Books']])
print("\nDétails des titres par catégorie :")
for index, row in df_categories.iterrows():
    print(f"\nCatégorie: {row['Category']}")
    print(f"Nombre de livres: {row['Number of Books']}")
    print("Titres des livres:")
    for title in row['Book Titles']:
        print(f"  - {title}")

# Sauvegarder en CSV, Excel et JSON
df_categories.to_csv('books_by_category_with_titles.csv', index=False)
df_categories.to_excel('books_by_category_with_titles.xlsx', index=False)
df_categories.to_json('books_by_category_with_titles.json', orient='records', lines=True)
```
