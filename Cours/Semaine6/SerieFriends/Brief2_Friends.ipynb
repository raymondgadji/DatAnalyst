# Saison 1

```python
import requests
from bs4 import BeautifulSoup
import csv

# URL de la page pour la saison 1
url = 'https://www.imdb.com/title/tt0108778/episodes/?season=1'

# Headers pour simuler une requête de navigateur
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Faire une requête GET
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Trouver tous les titres des épisodes
titles = soup.find_all("h4")

# Liste pour stocker les données
data = []

# Parcourir chaque titre
for i, h4 in enumerate(titles, start=1):
    # Titre complet
    title_text = h4.text.strip()

    # Parent pour récupérer les autres infos
    parent = h4.find_parent("div", class_="sc-5372d523-5 jBUStp")

    # Date de sortie
    span = parent.find("span")
    release_date = span.text.strip() if span else "Date non trouvée"

    # Description
    desc = parent.find_next("div", class_="ipc-html-content-inner-div")
    description = desc.text.strip() if desc else "Description non trouvée"

    # Note IMDb
    rating_tag = parent.find_next("span", class_="ipc-rating-star--rating")
    rating = rating_tag.text.strip() if rating_tag else "Note non trouvée"

    # Ajouter les données à la liste
    data.append({
        'Episode': f"Épisode {i}",
        'Titre': title_text,
        'Date': release_date,
        'Note IMDb': f"{rating}/10",
        'Description': description
    })

# Écrire les données dans un fichier CSV
with open('episodes_saison_1.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['Episode', 'Titre', 'Date', 'Note IMDb', 'Description'])
    writer.writeheader()
    writer.writerows(data)

print("Données exportées dans 'episodes_saison_1.csv'")
```

Ensuite convertir les variables selon les types

```python
import pandas as pd

# Liste des fichiers CSV
csv_files = [
    'episodes_saison_1.csv',
    'episodes_saison_2.csv',
    'episodes_saison_3.csv',
    'episodes_saison_4.csv',
    'episodes_saison_5.csv',
    'episodes_saison_6.csv',
    'episodes_saison_7.csv',
    'episodes_saison_8.csv',
    'episodes_saison_9.csv',
    'episodes_saison_10.csv'
]

# Liste pour stocker les DataFrames
dataframes = []

# Charger chaque CSV et ajouter une colonne 'Saison'
for i, file in enumerate(csv_files, start=1):
    df = pd.read_csv(file)
    df['Saison'] = f'Saison {i}'
    dataframes.append(df)

# Combiner tous les DataFrames en un seul
combined_df = pd.concat(dataframes, ignore_index=True)

# Conversion des types de données
# 1. Convertir "Episode" en int (supprimer "Épisode " et convertir)
combined_df['Episode'] = combined_df['Episode'].str.replace('Épisode ', '').astype(int)

# 2. Convertir "Note IMDb" en float (remplacer la virgule par un point et "/10")
combined_df['Note IMDb'] = combined_df['Note IMDb'].str.replace('/10', '').str.replace(',', '.').astype(float)

# 3. Convertir "Date" en datetime (gérer différents formats possibles)
combined_df['Date'] = pd.to_datetime(combined_df['Date'], errors='coerce', format='%a, %d %b %Y')

# 4. Convertir "Saison" en int (supprimer "Saison " et convertir)
combined_df['Saison'] = combined_df['Saison'].str.replace('Saison ', '').astype(int)

# Sauvegarder le DataFrame combiné dans un nouveau CSV
combined_df.to_csv('all_friends_episodes_cleaned.csv', index=False, encoding='utf-8')

# Afficher des informations sur le DataFrame
print("Shape du DataFrame combiné :", combined_df.shape)
print("\nInformations sur les variables :")
print(combined_df.info())

# Utilisation de describe() pour obtenir des statistiques descriptives
print("\nStatistiques descriptives :")
print(combined_df.describe(include='all'))
```
