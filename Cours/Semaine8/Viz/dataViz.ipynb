```python
df_bookWithCat = df_bookWithCat.rename(columns={'category': 'Category'})
df_bookWithCat
```

```python
# Merge the two df
df_merge = df_bookWithCat.merge(df_bookBycat, on='Category', how='left')
df_merge
```

```python
# display the first five rows
df_merge.head()
```

```python
# display the last five rows
df_merge.tail()
```

```python
# display how many values are in "Category" column
df_merge.Category.value_counts()
```

```python
# display how many NULL values are in "Category" column
# Both isna() and isnull() are identical functions 
# in pandas and will give you the same result when checking for NULL values.
na_Category_count = df_merge['Category'].isna().sum()
print(na_Category_count)
```

```python
# display how many NULL values are in "Category" column
# There is no difference in the count they produce; 
# the choice between them is a matter of preference or coding style.
null_Category_count = df_merge['Category'].isnull().sum()
print(null_Category_count)
```
